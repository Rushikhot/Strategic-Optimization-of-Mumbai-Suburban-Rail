```python
train['Station']=train['Station'].astype('category').cat.codes
train['Line']= train['Line'].astype('category').cat.codes
train['Station Code']=train['Station Code'].astype('category').cat.codes
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corre,
            annot=True,
            cmap='YlGnBu',
            linewidths=0.5,
            annot_kws={"size": 12, "weight": "bold"},
            fmt=".2f",
            cbar_kws={"shrink": 0.8,"pad": 0.05,"label": "Correlation Strength"})
plt.title('Correlation Heatmap of Mumbai Train Features', fontsize=14, weight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
